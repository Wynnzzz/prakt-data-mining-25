##### Grouping & Aggregating Data Frame

###### Grouping adalah suatu teknik memilah data pada Data Frame berdasarkan kriteria tertentu. Grouping memungkinkan untuk melakukan komputasi atau operasi tertentu kepada data-data spesifik tersebut.

```python
df_1.groupby('Meta Category')['Transaction Amount'].sum()
```

###### Aggregation digunakan untuk melakukan komputasi statistik terhadap data-data yang telah dikelompokkan

###### Operasi yang dapat diterapkan yaitu sum, min, max, mean, size, describe, first, last, count, std, var, sem

```python
df_1.groupby('Product Name').agg(rata_rata=('Transaction Amount', 'mean'),
                                 maksimum=('Transaction Amount', 'max'),
                                 minimum=('Transaction Amount', 'min'))
```

##### Data Merge

<img src="Img/types-of-joins.png">

```python
employees = pd.DataFrame({'id':[1,2,3,4],'name':['Johana','Mike','Patricia','James']})
employees
```

```python
salary = pd.DataFrame({'id':[1,2,3,5],'salary':[120000,100000,130000,90000]})
salary
```

```python
employees_with_salary = employees.merge(salary,
                                        left_on="id",
                                        right_on="id",
                                        how='inner')  # default
employees_with_salary
```

```python
employees_with_salary2 = employees.merge(salary,
                                        left_on="id",
                                        right_on="id")
employees_with_salary2
```

```python
employees_with_salary_outer = employees.merge(salary, 
                                              left_on='id',
                                              right_on='id', 
                                              how='outer')
employees_with_salary_outer
```

```python
employees_with_salary_left = employees.merge(salary,
                                             left_on='id',
                                             right_on='id',
                                             how='left')
employees_with_salary_left
```

```python
employees_with_salary_right = employees.merge(salary,
                                              left_on='id',
                                              right_on='id',
                                              how='right')
employees_with_salary_right
```

```python
employees.merge(salary, on="id")
```
